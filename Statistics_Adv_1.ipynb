**1) What is a random variable in probability theory?**

--> A random variable is a function that assigns a real number to each outcome in a sample space of a random experiment. Essentially, it provides a numerical representation of outcomes, allowing us to analyze and quantify the behavior of random phenomena.

**2) What are the types of random variables?**

--> 1) Discrete Random Variables:

These take on a countable or finite set of values.
Examples: Number of heads in 10 coin flips, number of cars passing through a toll booth in an hour, number of students attending a class.
Probability distributions are described by probability mass functions (PMFs).

2) Continuous Random Variables:

These can take on any value within a continuous interval or set of intervals.
Examples: Height of a person, temperature on a given day, time taken to complete a task.
Probability distributions are described by probability density functions (PDFs).

**3) What is the difference between discrete and continuous distributions?**

--> The main difference between discrete and continuous distributions lies in the type of outcomes they describe and how probabilities are assigned.

Discrete distributions deal with countable outcomes and assign probabilities directly to each outcome.

Continuous distributions deal with uncountably many outcomes, and probabilities are assigned to intervals via the density function, with the probability of any single exact value being zero.

**4) What are probability distribution functions (PDF)?**

--> A probability distribution function is a mathematical function that describes how the probabilities are distributed over the possible outcomes of a random variable. It provides a complete description of the likelihood of different outcomes occurring in a random experiment.

**5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

--> The PDF, often denoted as f(x), describes the relative likelihood of a continuous random variable X taking on a specific value x.

The CDF, denoted as F(x), gives the probability that the random variable X is less than or equal to a specific value x.

**6) What is a discrete uniform distribution?**

--> A discrete uniform distribution is a type of probability distribution where a finite or countably infinite set of outcomes are equally likely to occur. In other words, each possible value has the same probability of being selected.

**7) What are the key properties of a Bernoulli distribution?**

**10) What is a continuous uniform distribution?**

--> A continuous uniform distribution is a type of probability distribution where all outcomes in a specified interval are equally likely. In other words, if a random variable X is said to follow a continuous uniform distribution over the interval [a,b], then the probability density function (PDF) is constant across that interval.

**11) What are the characteristics of a normal distribution?**

--> A normal distribution, also known as a Gaussian distribution, has several key characteristics:

1) Bell-Shaped Curve: The graph of a normal distribution is symmetric and bell-shaped, centered around the mean.

2) Symmetry: The distribution is perfectly symmetrical about its mean, meaning the left and right sides are mirror images.

3) Mean, Median, and Mode are Equal: In a normal distribution, these three measures of central tendency are all located at the center of the distribution.

4) Defined by Two Parameters: It is characterized by the mean (μ), which determines the center, and the standard deviation (σ), which determines the spread or dispersion.

5) 68-95-99.7 Rule:

Approximately 68% of the data falls within one standard deviation (μ ± σ).
About 95% within two standard deviations (μ ± 2σ).
About 99.7% within three standard deviations (μ ± 3σ).

6) Asymptotic: The tails of the distribution approach, but never touch, the horizontal axis, extending infinitely in both directions.

**12) What is the standard normal distribution, and why is it important?**

--> The standard normal distribution is a specific type of normal (or Gaussian) distribution characterized by a mean of 0 and a standard deviation of 1. Its probability density function (PDF) is symmetric around the mean and follows the bell-shaped curve that is typical of normal distributions.

Why is it important?

1) Foundation for Statistical Inference: Many statistical methods assume normality because the normal distribution has well-understood properties. The standard normal distribution serves as a reference point for standardizing data.

2) Standardization (Z-scores): Any normal distribution can be converted to the standard normal distribution by subtracting the mean and dividing by the standard deviation. This process, called standardization, allows for easy comparison and calculation of probabilities across different normal distributions.

3) Lookup Tables and Probability Calculations: Standard normal distribution tables (z-tables) provide probabilities associated with z-scores. These tables are essential tools for calculating p-values, confidence intervals, and other statistical measures.

4) Central Limit Theorem: It states that, under certain conditions, the sum (or average) of a large number of independent random variables tends toward a normal distribution, often approximated by the standard normal distribution after standardization.

**13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

--> The Central Limit Theorem (CLT) is a fundamental principle in statistics that describes the behavior of the sampling distribution of the sample mean (or sum) when independent, identically distributed random variables are involved.

Why it is in statistics ?

1) Enables Approximate Inference:
The CLT allows statisticians to use the normal distribution as an approximation for the distribution of sample means, even when the original data is not normally distributed, provided the sample size is large enough.

2) Foundation for Hypothesis Testing and Confidence Intervals:
Many statistical procedures rely on the normality assumption of the sampling distribution, which the CLT justifies for large samples. This simplifies the computation of p-values, confidence intervals, and other inference tools.

3) Universal Applicability:
The CLT applies broadly across different types of data distributions, making it a versatile tool in empirical research, quality control, economics, and many other fields.

4) Simplifies Complex Problems:
By approximating complicated distributions with the normal distribution, it becomes easier to analyze and interpret data, especially in large-sample contexts.

**14) How does the Central Limit Theorem relate to the normal distribution?**

--> The Central Limit Theorem (CLT) is a fundamental principle in statistics that describes how the distribution of the sample mean approaches a normal distribution under certain conditions.

In relation to the normal distribution:

1) The CLT explains why the normal distribution is so prevalent in statistical inference. It shows that even if the underlying data is not normally distributed, the distribution of the sample mean will tend toward a normal distribution as the sample size grows large.

2) This property allows statisticians to make probabilistic statements about the sample mean using the well-understood properties of the normal distribution, such as calculating confidence intervals and conducting hypothesis tests.

**15) What is the application of Z statistics in hypothesis testing?**

--> Z statistics play a crucial role in hypothesis testing by providing a standardized way to determine how far a sample statistic (such as a sample mean or proportion) deviates from the hypothesized population parameter, under the assumption that the null hypothesis is true.

Application of Z Statistics in Hypothesis Testing:

1) Standardization of Test Statistics:
The Z statistic transforms the difference between the sample statistic and the population parameter into a standard normal variable. This standardization allows for comparison across different datasets and tests.

2) Testing Population Means (When Variance is Known):  

Scenario: When testing whether a sample mean differs from a known population mean, and the population variance is known.

**16) How do you calculate a Z-score, and what does it represent?**

--> A Z-score is a standardized value that indicates how many standard deviations an individual data point is from the mean of a dataset. It allows you to compare values from different distributions or datasets on a common scale.

How to calculate a Z-score:

Z = (X−μ) / σ

Z-score represents:

1. A Z-score tells you how far and in which direction X deviates from the mean, measured in standard deviations.
2. A Z-score of 0 indicates the data point is exactly at the mean.
3. A positive Z-score indicates the data point is above the mean.
4. A negative Z-score indicates it is below the mean.

**17) What are point estimates and interval estimates in statistics?**

--> Point Estimates
 A point estimate is a single numerical value that serves as the best guess or estimate of an unknown population parameter.

 Interval Estimates
 An interval estimate provides a range of plausible values for the population parameter, usually expressed with a confidence level.

**18) What is the significance of confidence intervals in statistical analysis?**

--> Confidence intervals (CIs) are fundamental tools in statistical analysis that provide a range of plausible values for an unknown population parameter (such as a mean, proportion, or difference between groups).

Their significance lies in several key aspects:

1) Quantifying Uncertainty:
Confidence intervals express the degree of uncertainty associated with an estimate. Instead of a single point estimate, a CI offers a range that likely contains the true parameter value.

2) Informing Decision-Making:
By providing a range rather than just a point estimate, confidence intervals help researchers, policymakers, and practitioners assess the precision and reliability of the results, aiding informed decisions.

3) Assessing Statistical Significance:
When comparing groups or testing hypotheses, if a confidence interval for the difference between groups does not include zero (or the null value), it suggests a statistically significant difference at the chosen confidence level.

4) Communicating Results Effectively:
Confidence intervals offer a more comprehensive picture of the data than p-values alone, allowing stakeholders to understand both the estimate and its potential variability.

**19) What is the relationship between a Z-score and a confidence interval?**

--> A Z-score and a confidence interval are closely related concepts in statistics, particularly in the context of estimating population parameters.

The Z-score acts as a critical value that determines how wide the confidence interval is for a given confidence level.

As the confidence level increases, the Z-score (critical value) increases, resulting in a wider interval.

The Z-score provides a standardized threshold used to account for variability and to ensure the specified confidence level.

**20) How are Z-scores used to compare different distributions?**

--> How Z-scores are used to compare different distributions:

Standardization of Data Points:
A Z-score is calculated using the formula:
Z = (X−μ) / σ

where X is the raw score, μ is the mean of the distribution, and σ sigmaσ is the standard deviation.

**21) What are the assumptions for applying the Central Limit Theorem?**

--> The Central Limit Theorem (CLT) states that, given a sufficiently large sample size, the sampling distribution of the sample mean will be approximately normally distributed, regardless of the underlying population distribution.

To apply the CLT correctly, certain assumptions should be met:

1) Independence of Observations: The data points must be independent of each other. Typically, this requires that the sample be randomly drawn from the population, and the observations do not influence one another.

2) Finite Variance: The population from which samples are drawn should have a finite variance. Infinite variance distributions (such as certain heavy-tailed distributions) do not satisfy the CLT conditions.

3) Identically Distributed: The observations should be identically distributed (i.i.d.), meaning each observation has the same probability distribution.

**22) What is the concept of expected value in a probability distribution?**

--> The concept of expected value in a probability distribution, often denoted as E[X] or μ, represents the average or mean value that a random variable X is expected to take over many repeated trials or observations.

**23) How does a probability distribution relate to the expected outcome of a random variable?**

--> A probability distribution provides a complete description of the likelihood of all possible outcomes of a random variable. It assigns probabilities to each outcome (discrete case) or provides a density function over possible values (continuous case).

The expected outcome of a random variable, known as its expected value or mean, is a key summary statistic derived from its probability distribution.

```python
## 3) Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

    def bernoulli_pmf(x, p):
      """
    Calculate the probability mass function (PMF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome, should be 0 or 1.
    p (float): The probability of success, 0 <= p <= 1.

    Returns:
    float: The probability of observing outcome x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1.")
    return p**x * (1 - p)**(1 - x)
```

```python
## 6) Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
## uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters of the discrete uniform distribution
a = 1  # lower bound (inclusive)
b = 10 # upper bound (inclusive)

# Generate the support of the distribution
x = np.arange(a, b + 1)

# Calculate the probability mass function (PMF)
pmf = np.ones_like(x) / (b - a + 1)

# Calculate the CDF
# For each point, the CDF is the sum of PMFs up to that point
cdf = np.cumsum(pmf)

# Plotting the CDF
plt.figure(figsize=(8, 4))
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.scatter(x, cdf, color='red')  # Mark the points
plt.title('CDF of a Discrete Uniform Distribution')
plt.xlabel('x')
plt.ylabel('CDF(x)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
## 14) Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
## distribution.

import numpy as np
import scipy.stats as stats

def calculate_z_score(data, data_point):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    data (list or array): The dataset.
    data_point (float): The data point to compute the Z-score for.

    Returns:
    float: The Z-score of the data point.
    """
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution.

    Parameters:
    z_score (float): The Z-score to compare.

    Returns:
    dict: Contains the percentile and the probability of observing a Z-score less than or equal to the given Z.
    """
    percentile = stats.norm.cdf(z_score)
    return {
        'percentile': percentile,
        'probability': percentile  # Probability of Z <= given Z-score
    }
```

```python
## 15) Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np

# Sample data
sample_data = np.array([50, 52, 48, 51, 49, 50, 53, 47, 49, 52])  # example data
mu_0 = 50  # hypothesized population mean
sigma = 2  # known population standard deviation
alpha = 0.05  # significance level

# Calculate sample mean and sample size
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Calculate standard error
standard_error = sigma / np.sqrt(n)

# Compute Z-statistic
z_score = (sample_mean - mu_0) / standard_error

# Determine the critical Z value for a two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Calculate p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Output the results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Statistic: {z_score}")
print(f"Critical Z-value at alpha={alpha}: ±{z_critical}")
print(f"P-value: {p_value}")

# Make a decision
if abs(z_score) > z_critical:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
## 16) Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

# Sample data: heights in cm
heights = [172, 168, 174, 169, 170, 171, 173, 175, 169, 172,
           168, 170, 171, 172, 173, 169, 170, 174, 172, 171,
           170, 169, 173, 174, 172, 171, 170, 169, 172]

# Calculate sample mean and standard deviation
sample_mean = np.mean(heights)
sample_std = np.std(heights, ddof=1)  # ddof=1 for sample standard deviation
n = len(heights)

# Choose confidence level
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the t critical value
t_crit = stats.t.ppf(1 - alpha/2, df=n-1)

# Compute the margin of error
margin_of_error = t_crit * (sample_std / np.sqrt(n))

# Calculate confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f} cm")
print(f"{int(confidence_level*100)}% Confidence Interval: ({lower_bound:.2f} cm, {upper_bound:.2f} cm)")
```

```python
## 19) Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parameters for the Poisson distribution
lambda_param = 3  # average rate (mean number of events)
k = 5             # value up to which we want the CDF

# Calculate the CDF at k
cdf_value = stats.poisson.cdf(k, mu=lambda_param)
print(f"CDF at k={k} for Poisson with λ={lambda_param}: {cdf_value:.4f}")

# Interpretation:
# The CDF at k=5 gives the probability that the number of events is less than or equal to 5.
# In other words, P(X ≤ 5).

# Optional: Plot the PMF and CDF
x = np.arange(0, 15)
pmf = stats.poisson.pmf(x, mu=lambda_param)
cdf = stats.poisson.cdf(x, mu=lambda_param)

plt.figure(figsize=(12, 6))

# Plot PMF
plt.subplot(1, 2, 1)
plt.bar(x, pmf, alpha=0.7, color='skyblue')
plt.title('Poisson PMF (λ=3)')
plt.xlabel('Number of events')
plt.ylabel('Probability')

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf, where='post', color='orange')
plt.xlabel('Number of events')
plt.ylabel('Cumulative Probability')
plt.title('Poisson CDF (λ=3)')

plt.tight_layout()
plt.show()
```

```python
## 21) Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

# Generate or input your datasets
# Example datasets:
dataset1 = np.random.normal(loc=0, scale=1, size=1000)
dataset2 = np.random.normal(loc=0, scale=2, size=1000)

# Calculate standard deviations
std1 = np.std(dataset1)
std2 = np.std(dataset2)

print(f"Standard Deviation of Dataset 1: {std1:.3f}")
print(f"Standard Deviation of Dataset 2: {std2:.3f}")

# Plot the distributions
plt.figure(figsize=(10, 6))
bins = 30

plt.hist(dataset1, bins=bins, alpha=0.5, label='Dataset 1')
plt.hist(dataset2, bins=bins, alpha=0.5, label='Dataset 2')

# Add lines for the mean and ±1 standard deviation
for dataset, std, label in zip([dataset1, dataset2], [std1, std2], ['Dataset 1', 'Dataset 2']):
    mean = np.mean(dataset)
    plt.axvline(mean, color='k', linestyle='--', linewidth=1)
    plt.axvline(mean + std, color='r', linestyle=':', linewidth=1, label=f'{label} +1 STD')
    plt.axvline(mean - std, color='r', linestyle=':', linewidth=1)

    plt.legend()
    plt.title('Comparison of Standard Deviations')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
## 22) Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

# Generate data
np.random.seed(0)  # for reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate quartiles
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

print(f"Range: {data_range}")
print(f"IQR: {IQR}")
```

```python
## 23) Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)
data = np.random.normal(loc=50, scale=20, size=100)  # mean=50, std=20

# Calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Z-score normalization
data_zscore = (data - mean) / std

# Plotting
plt.figure(figsize=(12, 6))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=15, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean:.2f}')
plt.legend()

# Normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(data_zscore, bins=15, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.axvline(0, color='blue', linestyle='dashed', linewidth=1, label='Mean = 0')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
## 24) Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
## distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution(mean=0, std_dev=1, size=1000):
    """
    Generate a dataset from a normal distribution and calculate skewness and kurtosis.

    Parameters:
    - mean: The mean of the normal distribution.
    - std_dev: The standard deviation of the normal distribution.
    - size: Number of data points to generate.
    Returns:
    - skewness: The skewness of the dataset.
    - kurtosis_value: The kurtosis of the dataset.
    """
    # Generate the dataset
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness
    skewness = skew(data)

    # Calculate kurtosis (Fisher’s definition: normal distribution -> 0)
    kurtosis_value = kurtosis(data)

    return skewness, kurtosis_value
```
